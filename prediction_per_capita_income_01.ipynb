```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
file_path = 'canada_per_capita_income.csv'
df = pd.read_csv(file_path)
df
```

```python
df.columns
```

```python
year = df[['year']]
per_capita = df[['per capita income (US$)']]
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per capita income (us$)')
plt.scatter(year, per_capita,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,per_capita )
```

```python
prediction = reg.predict([[2025]])
prediction
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per capita income (us$)')
plt.scatter(year, per_capita,color='red',marker='+')
plt.plot(year, reg.predict(year), color='blue')
```
